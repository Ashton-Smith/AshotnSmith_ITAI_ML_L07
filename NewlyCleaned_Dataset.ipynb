## Analyze Heart Disease Prevalence

### Subtask:
Analyze the 'target' column to determine the number and proportion of patients identified with or without heart disease.


**Reasoning**:
To determine the number and proportion of patients with and without heart disease, I will count the occurrences of each unique value in the 'target' column and then calculate their percentages.



```python
heart_disease_counts = df['target'].value_counts()
heart_disease_proportions = df['target'].value_counts(normalize=True) * 100

print("Heart Disease Prevalence (Counts):")
print(heart_disease_counts)
print("\nHeart Disease Prevalence (Proportions %):")
print(heart_disease_proportions)
```

## Visualize Potential Risk Factors

### Subtask:
Generate visualizations, such as histograms or box plots, to explore the distributions of key features and their relationships with the 'target' variable (heart disease), providing legends for clarity. This will help in identifying potential risk factors.


**Reasoning**:
The subtask requires generating visualizations to explore the relationship between key features and the 'target' variable. The first step is to import the necessary libraries, `matplotlib.pyplot` for plotting and `seaborn` for enhanced visualizations.



```python
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries imported successfully.")
```

**Reasoning**:
To visualize the relationship between key features and the 'target' variable, I will create a figure with subplots and generate box plots for 'age' and 'cholestoral' against 'target', and countplots for 'sex' and 'chest_pain_type' against 'target', ensuring proper labels, titles, and legends.



```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Box plot for 'age' vs 'target'
sns.boxplot(x='target', y='age', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution by Heart Disease Status')
axes[0, 0].set_xlabel('Heart Disease (0: No, 1: Yes)')
axes[0, 0].set_ylabel('Age')

# Box plot for 'cholestoral' vs 'target'
sns.boxplot(x='target', y='cholestoral', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Cholesterol Distribution by Heart Disease Status')
axes[0, 1].set_xlabel('Heart Disease (0: No, 1: Yes)')
axes[0, 1].set_ylabel('Cholesterol')

# Countplot for 'sex' vs 'target'
sns.countplot(x='sex', hue='target', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Sex Distribution by Heart Disease Status')
axes[1, 0].set_xlabel('Sex')
axes[1, 0].set_ylabel('Count')
axes[1, 0].legend(title='Heart Disease', labels=['No', 'Yes'])

# Countplot for 'chest_pain_type' vs 'target'
sns.countplot(x='chest_pain_type', hue='target', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Chest Pain Type Distribution by Heart Disease Status')
axes[1, 1].set_xlabel('Chest Pain Type')
axes[1, 1].set_ylabel('Count')
axes[1, 1].legend(title='Heart Disease', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()
```
